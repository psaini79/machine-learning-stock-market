```python
class StdOutListener(StreamListener):

    def on_data(self, data):
        data = json.loads(data)
        #print (data)
        #return True        
        if data.get("lang") != "en": 
            return True
        try:
            timestamp = data['timestamp_ms']            
            # Get longer 280 char tweets if possible
            if data.get("extended_tweet"):
                tweet = data['extended_tweet']["full_text"]
            else:
                tweet = data["text"]
            url = "https://www.twitter.com/i/web/status/" + data["id_str"]
            #print("url: ", url.encode("utf-8"))
            user = data["user"]["screen_name"]
            #print(user)
            verified = data["user"]["verified"]
            #print(verified)
            write_to_csv([timestamp.encode("utf-8"), tweet.encode("utf-8"), user.encode("utf-8"), verified, url.encode("utf-8")])
            #print(tweet.encode("utf-8"))
             
        except KeyError as e:
            print("Keyerror:", e)

        return True

    def on_error(self, status):
        print ("Status:", status)
```

```python
def clean_tweet(tweet):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t]) | (\w+:\/\/\S+)", " ", tweet).split())
```

```python
def get_tweet_sentiment(tweet):
    analysis = TextBlob(clean_tweet(tweet))
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity == 0: 
        return 'neutral'
    else: 
        return 'negative'
```

```python
tweets_data_path = '../tweet.csv'
tweets_data = []
tweets_file = open(tweets_data_path, "r")
for tweet in tweets_file:
    if(re.match("timestamp", tweet)):
        continue
    try:
        parsed_tweet = {}        
        parsed_tweet['text'] = tweet
        parsed_tweet['sentiment'] = get_tweet_sentiment(tweet)
        tweets_data.append(parsed_tweet)
        """
        if tweet.retweet_count > 0:
            if parsed_tweet not in tweets:
                tweets_data.append(parsed_tweet)
            else:
                tweets_data.append(parsed_tweet)
        """
    except tweepy.TweepError as e:
        print("Error : " + str(e)) 
tweets_file.close()
        
```
